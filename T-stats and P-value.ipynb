```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with the given data
data = {'Group': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
        'Score': [85, 87, 82, 90, 88, 78, 80, 79, 83, 81]}

df = pd.DataFrame(data)

# Separate the data into two groups
group_a = df[df['Group'] == 'A']
group_b = df[df['Group'] == 'B']



# Plot histograms for both groups
plt.hist(group_a['Score'], alpha=0.5, label='Group A', bins=5)
plt.hist(group_b['Score'], alpha=0.5, label='Group B', bins=5)

plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()


print("This visualization will help us observe the spread of scores in both groups and any potential outliers.")
```

```python
#1.2
```

```python
import numpy as np

mean_a = np.mean(group_a['Score'])
std_dev_a = np.std(group_a['Score'])

mean_b = np.mean(group_b['Score'])
std_dev_b = np.std(group_b['Score'])

n1 = len(group_a)
n2 = len(group_b)

print(f"Group A - Mean: {mean_a}, Standard Deviation: {std_dev_a}")
print(f"Group B - Mean: {mean_b}, Standard Deviation: {std_dev_b}")
print("We can see that standard deviation of Group B students is less than that of group A students which signifies that values of Group B are closely related than that of Group A")
```

```python
#1.3

import math
from scipy.stats import norm
from scipy import stats

t_stat, p_value = stats.ttest_ind(group_a['Score'], group_b['Score'])

#t = (mean_a - mean_b) / math.sqrt((std_dev_a**2 / n1) + (std_dev_b**2 / n2))
print(f"t-statistic: {t_stat}, p-value: {p_value}")
#print('t : ',t)

print("""\nWe can conclude that Group A students grasp more than that of Group B, as mean of Group A is more than Group B 
students and p-value is less than 0.005.""")
```

```python
#2

import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with the given data
data = {
    'Day': [1, 2, 3, 4, 5, 6, 7],
    'Control Group (Hours)': [7.5, 7.8, 7.6, 7.7, 7.9, 7.8, 7.6],
    'Treatment Group (Hours)': [8.2, 8.4, 8.1, 8.3, 8.5, 8.2, 8.4]
}

df = pd.DataFrame(data)

# Plot the data
plt.plot(df['Day'], df['Control Group (Hours)'], label='Control Group', marker='o')
plt.plot(df['Day'], df['Treatment Group (Hours)'], label='Treatment Group', marker='o')

plt.xlabel('Day')
plt.ylabel('Average Sleep Duration (Hours)')
plt.legend()
plt.title('Average Sleep Duration for Control and Treatment Groups')
plt.grid()
plt.show()

print("""\nThe treatment group generally has a higher average sleep duration than the control group.""")
```

```python
#2.2
mean_control = df['Control Group (Hours)'].mean()
mean_treatment = df['Treatment Group (Hours)'].mean()

print(f"Control Group - Mean Sleep Duration: {mean_control} hours")
print(f"Treatment Group - Mean Sleep Duration: {mean_treatment} hours")
```

```python
#2.3
from scipy import stats

t_stat, p_value = stats.ttest_rel(df['Control Group (Hours)'], df['Treatment Group (Hours)'])
print(f"t-statistic: {t_stat}, p-value: {p_value}")

#2.4
print("""\nIf the p-value is less than 0.05, it means that there 
is a significant difference in sleep duration between the control
and treatment groups. This suggests that the drug has a significant effect on increasing 
sleep duration.""")

#2.5
print("""\nYes, based on the analysis and the significance of the p-value, it 
appears that the drug is effective in increasing the duration of 
sleep. The paired t-test indicates a statistically significant 
difference in sleep duration between the treatment group and the 
control group. The drug has a positive effect on increasing sleep duration.""")
```
